```python
import re
import pandas as pd
import os

# === 1️⃣ Gather all .txt files ===
periods = ["1850_2100", "1850_1900", "1900_1950", "2000_2050", "1950_2000"]
epochs = ["30", "60", "80", "100"]
periods2 = ["1850-2100", "1850-1900", "1900-1950", "2000-2050", "1950-2000"]

files = [
    f"/home/sc.uni-leipzig.de/fl53wumy/llaae_new/DistributionalPrincipalAutoencoder/"
    f"dpa_results_analysis/ETH_analysis_results/final_analysis_test_ETH/"
    f"model_trained_for_{no_epochs}_epochs/period_{period}/log_metrics_{period2}.txt"
    for no_epochs in epochs
    for period, period2 in zip(periods, periods2)
    if os.path.exists(
        f"/home/sc.uni-leipzig.de/fl53wumy/llaae_new/DistributionalPrincipalAutoencoder/"
        f"dpa_results_analysis/ETH_analysis_results/final_analysis_test_ETH/"
        f"model_trained_for_{no_epochs}_epochs/period_{period}/log_metrics_{period2}.txt"
    )
]

print(f"✅ Found {len(files)} existing files:")
for f in files:
    print(f)


# === 2️⃣ Regex patterns ===
patterns = {
    "Factual mean diff": r"Factual mean quantile differences \(across all quantiles\): ([\d\.\-eE]+)",
    "Factual 0.05 diff": r"Factual 0\.05-quantile differences: ([\d\.\-eE]+)",
    "Counterfactual mean diff": r"Counterfactual mean quantile differences \(across all quantiles\): ([\d\.\-eE]+)",
    "Counterfactual 0.05 diff": r"Counterfactual 0\.05-quantile differences: ([\d\.\-eE]+)",
    "Factual auto diff": r"Spatial mean factual autocorrelation difference: ([\d\.\-eE]+)",
    "Counterfactual auto diff": r"Spatial mean counterfactual autocorrelation difference: ([\d\.\-eE]+)",
    "Factual Energy-Score": r"Factual Energy-Score spatial mean: ([\d\.\-eE]+)",
    "Factual S1-Score": r"Factual S1-Score spatial mean: ([\d\.\-eE]+)",
    "Factual S2-Score": r"Factual S2-Score spatial mean: ([\d\.\-eE]+)",
    "Counterfactual Energy-Score": r"Counterfactual Energy-Score spatial mean: ([\d\.\-eE]+)",
    "Counterfactual S1-Score": r"Counterfactual S1-Score spatial mean: ([\d\.\-eE]+)",
    "Counterfactual S2-Score": r"Counterfactual S2-Score spatial mean: ([\d\.\-eE]+)",
    "Factual RI": r"Factual RI spatial mean: ([\d\.\-eE]+)",
    "Counterfactual RI": r"Counterfactual RI spatial mean: ([\d\.\-eE]+)",
}


# === 3️⃣ Parse all files ===
data = []
for file in files:
    with open(file, "r") as f:
        content = f.read()

    # Extract period and epoch info
    epoch_match = re.search(r"model_trained_for_(\d+)_epochs", file)
    period_match = re.search(r"period_(\d{4}_\d{4})", file)
    period_pretty = period_match.group(1).replace("_", "–") if period_match else "unknown"
    epoch = epoch_match.group(1) if epoch_match else "unknown"

    # Create readable entry name
    row = {"Entry": f"{period_pretty} ({epoch} epochs)"}

    # Extract all metrics using regex
    for key, pattern in patterns.items():
        match = re.search(pattern, content)
        row[key] = float(match.group(1)) if match else None
    data.append(row)


# === 4️⃣ Convert to DataFrame ===
df = pd.DataFrame(data)


# === 5️⃣ Split into factual / counterfactual ===
factual_cols = ["Entry"] + [c for c in df.columns if c.startswith("Factual")]
counterfactual_cols = ["Entry"] + [c for c in df.columns if c.startswith("Counterfactual")]

df_factual = df[factual_cols].copy()
df_counter = df[counterfactual_cols].copy()


# === 6️⃣ Clean and rename columns ===
df_factual.columns = [c.replace("Factual ", "") for c in df_factual.columns]
df_counter.columns = [c.replace("Counterfactual ", "") for c in df_counter.columns]

rename_map = {
    "mean diff": "Quantiles mean difference",
    "0.05 diff": "Quantile\n0.05 difference",
    "auto diff": "Autocorrelation\nmean difference",
    "Energy-Score": "Spatial mean\nEnergy Score",
    "S1-Score": "Spatial mean\nS1-score",
    "S2-Score": "Spatial mean\nS2-score",
    "RI": "Spatial mean RI"
}
df_factual.rename(columns=rename_map, inplace=True)
df_counter.rename(columns=rename_map, inplace=True)


# === 7️⃣ Highlight rules: bold minima, absolute-min for 3rd column, italicize 1850–2100 ===
def format_markdown(df):
    """Return formatted DataFrame with:
       - min values bolded (per column)
       - 3rd column bolds the value with min absolute value
       - entries spanning 1850–2100 italicized
    """
    df_fmt = df.copy()

    # Detect numeric columns
    numeric_cols = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c])]

    for col in numeric_cols:
        if "Autocorrelation" in col:
            # ✅ Special rule: bold minimum *absolute* value
            min_abs_val = df[col].abs().min()
            df_fmt[col] = df[col].apply(
                lambda x: f"**{x:.4f}**" if abs(x) == min_abs_val else f"{x:.4f}" if pd.notnull(x) else ""
            )
        else:
            # Normal min rule
            min_val = df[col].min()
            df_fmt[col] = df[col].apply(
                lambda x: f"**{x:.4f}**" if x == min_val else f"{x:.4f}" if pd.notnull(x) else ""
            )

    # Italicize entries containing "1850–2100"
    df_fmt["Entry"] = df_fmt["Entry"].apply(
        lambda e: f"*{e}*" if "1850–2100" in e else e
    )

    return df_fmt


df_factual_fmt = format_markdown(df_factual)
df_counter_fmt = format_markdown(df_counter)

# Convert to Markdown
factual_md = df_factual_fmt.to_markdown(index=False)
counter_md = df_counter_fmt.to_markdown(index=False)


# === 8️⃣ Print & save ===
print("### Factual Results\n")
print(factual_md)
print("\n\n### Counterfactual Results\n")
print(counter_md)

with open("summary_tables.md", "w") as f:
    f.write("### Factual Results\n\n")
    f.write(factual_md + "\n\n")
    f.write("### Counterfactual Results\n\n")
    f.write(counter_md + "\n")

print("✅ Markdown summary saved as summary_tables.md")
```
