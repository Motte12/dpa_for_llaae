```python
import matplotlib.pyplot as plt
import numpy as np

# A4 size in inches (8.27 x 11.69)
fig, axs = plt.subplots(2, 2, figsize=(8.27, 11.69))  # 2x2 grid of subplots

# Create some example plots
x = np.linspace(0, 10, 100)

# Plot 1
axs[0, 0].plot(x, np.sin(x), label="sin(x)")
axs[0, 0].set_title("Plot 1: Sine Wave")
axs[0, 0].legend()

# Plot 2
axs[0, 1].plot(x, np.cos(x), label="cos(x)", color='orange')
axs[0, 1].set_title("Plot 2: Cosine Wave")
axs[0, 1].legend()

# Plot 3
axs[1, 0].scatter(x, np.tan(x), label="tan(x)", color='green')
axs[1, 0].set_title("Plot 3: Tangent")
axs[1, 0].legend()

# Plot 4
axs[1, 1].bar([1, 2, 3, 4], [5, 6, 7, 8], label="Bar Chart", color='red')
axs[1, 1].set_title("Plot 4: Bar Chart")
axs[1, 1].legend()

# Adjust spacing to prevent overlap
plt.tight_layout()

# Save to file (optional)
#plt.savefig('multiple_plots_a4.pdf')  # Can save as PDF for printing

plt.show()
```
